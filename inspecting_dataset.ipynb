```python
from ogb.graphproppred import GraphPropPredDataset

dataset = GraphPropPredDataset(name = "ogbg-molhiv")

split_idx = dataset.get_idx_split()
train_idx, valid_idx, test_idx = split_idx["train"], split_idx["valid"], split_idx["test"]

### set i as an arbitrary index
graph, label = dataset[1] # graph: library-agnostic graph object
```

```python
graph_lengths = [graph['num_nodes'] for graph, _ in dataset]
max_length_index = graph_lengths.index(max(graph_lengths))
print(max_length_index)
```

```python
graph, label = dataset[35616]
num_nodes = graph['num_nodes']
num_nodes
```

```python
num_edges = graph['edge_index'].shape[1]
num_edges
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Assuming dataset is already loaded and contains the graph data
graph_tuple = dataset[35616]

# Access the dictionary within the tuple
graph = graph_tuple[0]  # Adjust the index if the dictionary is not the first element

# Extracting data
edge_index = graph['edge_index']
num_nodes = graph['num_nodes']

# Create a networkx graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(range(num_nodes))

# Add edges
edges = list(zip(edge_index[0], edge_index[1]))
G.add_edges_from(edges)

# Plot the graph
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Network Graph")
plt.show()
```
